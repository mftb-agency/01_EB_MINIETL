PROYECTO MINI ETL

```python
# librerias para el proyecto
import pandas as pd
import numpy as np
```

EXTRAER

```python
# Cargar el archivo CSV
df = pd.read_csv("ventas.csv")

# Ver las primeras filas
df.head()
```

TRANSFORMAR

```python
# Crear una nueva columna de total por venta
df["total_venta"] = df["unidades"] * df["precio_unitario"]

# Agrupar por tienda y sumar las ventas
ventas_por_tienda = df.groupby("tienda")["total_venta"].sum().reset_index()

# Renombrar columnas
ventas_por_tienda.columns = ["Tienda", "Total Ventas"]

# Ordenar de mayor a menor
ventas_por_tienda = ventas_por_tienda.sort_values("Total Ventas", ascending=False)

# Mostrar resultados
ventas_por_tienda
```

CARGAR

```python
# Exportar el archivo Excel
ventas_por_tienda.to_excel("reporte_final.xlsx", index=False)

print("隆Reporte generado con 茅xito!")
```
